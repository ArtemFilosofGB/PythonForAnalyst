## Задание 1.
1. Скачать данные по ссылке
https://www.kaggle.com/datasets/ionaskel/laptop-prices
2. Считать данные с помощью pandas
3. Вывести на экран первые 5 строк
4. Создать новый признак Cpu_Company, который будет
содержать только название фирмы, которая произвела CPU
5. Создать новый признак Memory_Amount, который будет
содержать только количество Gb памяти без указания типа
носителя
6. Создать новый признак Memory_Type, который будет
содержать только тип носителя (HDD/SDD/др.)
7. Удалите признаки Memory и ScreenResolution

```python
import pandas as pd


df = pd.read_csv("laptops.csv", sep=';')
df.tail()
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
df.head(20)
```

```python
def convert_to_gb(x):
  memory = x.split()[0]
  if memory.endswith("GB"):
    res = memory.replace("GB", "")
  elif memory.endswith("TB"):
    res = float(memory.replace("TB", "")) * 1024
  return int(res)


df['Memory_Amount'] = df['Memory'].apply(lambda x: convert_to_gb(x))
df.tail()
```

```python
def convert_to_type(x):
  return x.replace(" ", '').split("B")[-1]


df['Memory_Type'] = df['Memory'].apply(lambda x: convert_to_type(x))
df.head()
```

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.head()
```
